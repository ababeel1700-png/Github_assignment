```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset("titanic")
kashti.head()
```

```python
sns.boxplot(x="sex", y = "fare", data=kashti)
```

```python
p1 = sns.boxplot(x="survived", y="age", showmeans=True, data=kashti)
plt.show()
```

```python
sns.boxplot(x="survived",
            y="age",
            showmeans=True,
            meanprops={
                "marker":"+", 
                    "markersize":"12",
                    "markeredgecolor":"red"},
            data=kashti)
plt.show()
```

```python
sns.boxplot(x="survived",
            y="age",
            showmeans=True,
            meanprops={
                "marker":"+", 
                    "markersize":"12",
                    "markeredgecolor":"red"},
            data=kashti)
#show labels
plt.xlabel("How many survived")
plt.ylabel("age (years)")
plt.title("Box plot of how many drowned", size=14, weight='bold')
plt.show()
```
