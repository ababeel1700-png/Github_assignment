```python
import seaborn as sns
import matplotlib.pyplot as plt

kashti=sns.load_dataset("titanic")
kashti
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female", "male"], ci=None, palette="dark:grey")
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


kashti=sns.load_dataset("titanic")
sns.barplot(x="sex", y="alone", hue="who", data=kashti, estimator =np.median)
plt.show()
```
